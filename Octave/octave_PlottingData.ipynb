```octave
t=[0:0.01:0.98]
```

```octave
y1=sin(2*pi*4*t)
```

```octave
graphics_toolkit ("gnuplot"); % 주피터에서 plot하기위함
```

```octave
plot(t,y1); %그래프는 CLI에서 확인해야할듯
```

```octave
y2=cos(2*pi*4*t)
```

```octave
plot(t,y2)
```

위의 사인함수와 코사인함수를 각각이 아닌 한번에 출력하려면

```octave
plot(t,y1);
hold on;
plot(t,y2);
```

```octave
% 다양한 설정
plot(t,y1);
hold on;
plot(t,y2);
xlabel('time')
ylabel('value')
legend('sin','cos')
title('My plot')
print -dpng 'myPlot.png' %png로 저장
```

octave는 다른 그래프를 지정할 수 있다

```octave
figure(1); plot(t,y1);
```

```octave
figure(2); plot(t,y2);
```

```octave
%화면분할
subplot(1,2,1); %Divides plot a 1x2 grid, access first element
plot(t,y1);
subplot(1,2,2);
plot(t,y2,'r'); %r은 색깔 지정. 그냥해봄
```

```octave
%축의 범위
subplot(1,2,1);
plot(t,y1);
subplot(1,2,2);
plot(t,y1,'g');
axis([0.4 0.6 -1 -0.5])
```
